# merge 메서드 사용하기

```python
person = pd.read_csv('pandas_data/survey_person.csv') 
site = pd.read_csv('pandas_data/survey_site.csv') 
survey = pd.read_csv('pandas_data/survey_survey.csv') 
visited = pd.read_csv('pandas_data/survey_visited.csv')
```

### 1.site의 name 칼럼이랑 visited의 site칼럼이랑 merge

```python
visited['name']=visited['site']
visited1=visited[['name','ident','dated']]
visited1
```

```python
visit_site = pd.merge(site,visited1, on='name')
visit_site
```

### 2. person의 ident칼럼과  survey의 person칼럼 merge


```python
person
```

```python
survey['ident']=survey['person']
survey1=survey[['ident','taken','quant','reading']]
survey1
```

```python
person_survey=pd.merge(person,survey1, on='ident')
person_survey
```
