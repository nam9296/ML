# 갭마인더 데이터 집합 불러오기

```python
import pandas as pd
df = pd.read_csv('pandas_data/gapminder.tsv', sep='\t')
df
```

# 불러온 데이터 집합 살펴보기

```python
# 타입, shape, info,...
df.shape
df.info()
```

# 열 단위로 데이터 추출하기

```python
df[['country']]
df[['continent']]
df[['year']]
df[['lifeExp']]
df[['pop']]
df[['gdpPercap']]
```

# loc 속성으로 행 단위 데이터 추출하기

```python
df.loc[:,['year','pop']]
```

```python
df.iloc[:,[2,4]]
```

# tail과 loc는 조금 달라요!

```python
type(df.tail())
```

```python
type(df.loc[1])
```

# iloc 속성으로 행 단위 데이터 추출하기

```python
df.iloc[:15,0]
```

# 파이썬 슬라이싱 구문을 조합하여 원하는 데이터 추출하기

```python
df.iloc[1:4,[3]]
```

# iloc 속성과 range 메서드로 원하는 데이터 추출하기

```python
df.iloc[:,list(range(3))]
```

# 열 지정값에 파이썬 슬라이싱을 사용하여 원하는 데이터 추출하기

```python
df.iloc[:,1:4]
```

# loc, iloc 자유자재로 사용하기

```python
df.iloc[[1,11,22],[2,5]]
```

# 그룹화한 데이터의 평균 구하기

```python
# 국가별, 대륙별, 연도 기준으로 그룹화 -> 연산
df.groupby('country').mean()
df.groupby('continent').mean()
df.groupby('year').mean()
```

# 그룹화한 데이터의 개수 세어보기

```python
df.groupby('year').count()
```

```python
df.groupby('year').size()
```
