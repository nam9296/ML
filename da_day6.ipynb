```python
vs=visited.loc[[0,2,6],]
vs
```

```python
# 사이트의 name열과 vs변수에 있는 site값이 같으므로 하나로 합칠수 있을것 같다.
```

```python
# site가 left vs가 right
# site에서 name 칼럼, vs에서 site칼럼
sv=site.merge(vs, left_on='name', right_on='site')
sv
```

```python
sv=site.merge(visited, left_on='name', right_on='site')
sv
```

```python
sp=plt.figure()
axes1=sp.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip']) # scater : 산점도 그래프
axes1.set_title("total_bill vs tip") # 제목 정하기
axes1.set_xlabel("total bill") 
axes1.set_ylabel("tip") 
```

```python
boxp=plt.figure()
axes1=boxp.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
             tips[tips['sex']=='Male']['tip']]) 
axes1.set_title("sex") # 제목 정하기
axes1.set_xlabel("tip") 
axes1.set_ylabel("boxplot") 
```

```python
def cs(a):
    if a == 'Femail':
        return 0
    else:
        return 1
```

```python
# cs함수에 sex열 전달
tips['sc']=tips['sex'].apply(cs)
```

```python
tips
```

```python
sp=plt.figure()
axes1=sp.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sc'],
    alpha=0.5
    
) 
axes1.set_title("scatter") 
axes1.set_xlabel("total_bill") 
axes1.set_ylabel("tip") 
```
